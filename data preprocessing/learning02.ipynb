```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading Data
dataset = pd.read_csv("./drive/MyDrive/Machine Learning/Data Preprocessing/data/data.csv")
print(dataset)
```

```python
# Find & Remove Outliers(the absolute unique/exceptional ones)
# Plotting the data
plt.hist(dataset['Age'], bins=len(dataset)-1)
plt.show()
print()

# Quantile
lowerlimit = dataset["Age"].quantile(0.05)
print(lowerlimit)
dataset[dataset["Age"] < lowerlimit]

upperlimit = dataset["Age"].quantile(0.95)
print(upperlimit)
dataset[dataset["Age"] > upperlimit]

dataset = dataset[(dataset["Age"] > lowerlimit) & (dataset["Age"] < upperlimit)]
print(dataset)
```
