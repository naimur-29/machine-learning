# Data Preprocessing

### Handling Null Values
- Droping data with null vaues
- Replacing the null with mean or median

```python
import pandas as pd
```

#### Doping the missing values

```python
# load data
dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning/rfsn_vhai_data/data/penguins.csv")
print("\nBEFORE REMOVING NULLS")
print(dataset)

# check null values
print("\nCHECKING NULLS")
print(dataset.isnull().sum())

# drop
dataset.dropna(inplace=True)
print("\nAFTER REMOVING NULLS")
print(dataset)

# check null values
print("\nCHECKING NULLS")
print(dataset.isnull().sum())
```

#### Replacing null values with mean if the scale is normal

```python
#load dataset
dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning/rfsn_vhai_data/data/penguins.csv")
print("\nBEFORE REPLACING...")
print(dataset)


# replacing process
import numpy as np
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(dataset.iloc[:, 2:6])
# doesn't work with non numeric values so:
dataset.iloc[:, 2:6] = imputer.transform(dataset.iloc[:, 2:6])
print("\nAFTER REPLACING WITH MEAN...")
print(dataset)
```

#### Replacing null with median if the scale is too high or too low

```python
# load dataset
dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning/rfsn_vhai_data/data/penguins.csv")
print("\nBEFORE REPLACING...")
print(dataset)


# replacing process:
import numpy as np
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='median')
# doesn't work with non number values so:
imputer.fit(dataset.iloc[:, 2:6])
dataset.iloc[:, 2:6] = imputer.transform(dataset.iloc[:, 2:6])
print("\nAFTER REPLACING...")
print(dataset)
```

### Data Encoding
- OneHotEncoder
- LabelEncoder

#### preparation

```python
# data loading
dataset = pd.read_json("/content/drive/MyDrive/Machine Learning/rfsn_vhai_data/data/penguins.json")
dataset
```

```python
#checking if there's null
print(dataset.isnull().sum())

# handling null values:
import numpy as np
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(dataset.iloc[:, 2:6])
dataset.iloc[:, 2:6] = imputer.transform(dataset.iloc[:, 2:6])

# just in case if there's still some NaN amongs non numerics:
dataset.dropna(inplace=True)
dataset
```

```python
# splitting dependent(y) and independent(x) vecots
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

print("\nINDEPENDENT DATA:")
print(x)

print("\nDEPENDENT DATA:")
print(y)
```

#### OneHotEncoder Usually The Independant Vecotrs, also data that are categorical

```python
print("\nBEFORE ENCODING:")
print(x[0])

# Encoding:
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

column_encoder = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), list(range(2)))], remainder='passthrough')
x = column_encoder.fit_transform(x)
print("\nAFTER ENCODING:")
print(x[0])
```

#### Label Encoding (usually the dependent / y value)

```python
print("\nBEFORE:")
print(y)

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
print("\nAFTER:")
print(y)
```

### Feature Scaling

```python
print("\nBEFORE:")
print(x[0])

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

# transformer = ColumnTransformer(transformers=[('scaler', StandardScaler(), list(range(6, 10)))], remainder='passthrough')
scaler = StandardScaler()
x[:, -4:] = scaler.fit_transform(x[:, -4:])
print("\nAFTER:")
print(x[0])
```
