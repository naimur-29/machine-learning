```python
import pandas as pd
```

```python
#Loading Data
df = pd.read_csv("./drive/MyDrive/Machine Learning/Data Preprocessing/data/duplicate-data.csv")
df
```

```python
# Drop Duplicates
df.drop_duplicates()
```

```python
# Drop Duplicates with subset
df.drop_duplicates(subset=['Company'])
```

```python
# Multiple subsets/columns
df.drop_duplicates(subset=['body style', 'fuel type'])
```

```python
# keep method (default = fast)
df.drop_duplicates(subset=['body style', 'fuel type'], keep='last')
```
