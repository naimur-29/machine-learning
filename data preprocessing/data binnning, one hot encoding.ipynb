```python
import pandas as pd
import numpy as np
```

```python
cricketers = {
    'name': ['Virat', 'Rohit', 'Babar', 'Boult', "Bumrah", "Mujeeb", 'Nabi', 'Strokes'],
    'age': [31, 33, 25, 31, 26, 19, 35, 29],
    'grade': ['Best', "Better", "Good", 'Better', 'Good', 'Best', 'Better', 'Good'],
    'role': ['Batsman', 'Batsman', 'Batsman', 'Bowler', "Bowler", 'Bowler', 'All-Rounder', "All-Rounder"],
    'rating': [821, 855, 829, 722, 719, 701, 301, 285],
    'country': ["IND", 'IND', 'PAK', "NZ", 'IND', 'AFG', 'AFG', 'ENG']
}
```

```python
cricketers = pd.DataFrame(cricketers)
cricketers
```

```python
# Data Binning (categorizing different data)
def binningFunction(col, cut_points, labels=None):
  minval = col.min()
  maxval = col.max()
  break_points = [minval] + cut_points + [maxval]
  print(break_points)
  if not labels:
    labels = range(len(cut_points) + 1)
  colBin = pd.cut(col, bins=break_points, labels=labels, include_lowest=True)
  return colBin

cut_points = [25, 30]
labels = ['young', 'senior', 'super-senior']
cricketers['Exp_category'] = binningFunction(cricketers['age'], cut_points, labels)
cricketers
```

```python
# One Hot Encoding with pandas
columns  = ['role']
dummy_data = pd.get_dummies(cricketers[columns])
cricketers = pd.concat([cricketers, dummy_data], axis=1)
cricketers
```

```python
cricketers = cricketers.drop(columns, axis=1)
cricketers
```
