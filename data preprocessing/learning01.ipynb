```python
import numpy as np
import pandas as pd
```

```python
#Loading Data
dataset = pd.read_csv("./drive/MyDrive/Machine Learning/Data Preprocessing/data/data.csv")
print(dataset)
```

```python
#Create dependent & independent variable vectors
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:, -1].values
print(x)
print(y)
```

```python
#Handle missing data
#Count the number of missing vlues in each column
# print(dataset.isnull().sum())

# 1.Drop the records
dataset.dropna(inplace=True)
# print(dataset)

# 2.Replace the missing values
print(x, '\n')
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
print(x)
```

```python
# Data Encoding
# Handle/Encode categorical data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

column_encoder = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(column_encoder.fit_transform(x))
print(x, '\n')

# Label Encoding
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
print(y)
```

```python
# Split the dataset for training & testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=9)
print(x_train, y_train, '\n')
print(x_test, y_test)
```

```python
# Feature Scaling - Standardiztion & Normalization
# Scaling everything to the same range

# Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train[:, 4:] = scaler.fit_transform(x_train[:, 4:])
x_test[:, 4:] = scaler.fit_transform(x_test[:, 4:])
print(x_train, '\n')
print(x_test)
```
